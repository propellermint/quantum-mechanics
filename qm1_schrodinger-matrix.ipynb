```python
import numpy as np
import matplotlib.pyplot as plt

from jupyterthemes import jtplot
jtplot.style(theme='oceans16', context='notebook', ticks=True, grid=False)
```

```python
'''Формирование матрицы гамильтониана.Сначала рассматривается
свободное движение (V = 0)'''
size = 20


def potential(n):
    if n == 50: return -1
    else: return 0
    
    
def f(i,j):
    if i == j: return 2.0 + potential(i)
    elif i - j == 1: return -1.0
    elif j - i == 1: return -1.0
    else: return 0.0
```

```python
# Оформление матрицы
h = [[f(i,j) for j in range(size)] for i in range(size)]
h = np.array(h)
print(h)
```

```python
#  Compute the eigenvalues and right eigenvectors of a square array.
eig, eve = np.linalg.eig(h)
eig  #  Выводит массив собственных значений
```

```python
plt.figure(figsize=(15, 10))
plt.scatter([i for i in range(size)], sorted(eig), s=3)
plt.ylabel('E_n')
plt.xlabel('n')
plt.axhline(y = 0, color="r")
plt.show()
```

```python
'''(*Пример существования локального уровня:*)'''

size = 100
# l - номер локального уровня

def pot(n, l):
    if n == l: return -1
    else: return 0
    
    
def ff(i, j, l):
    if i == j: return 2.0 + pot(i, l)
    elif i - j == 1: return -1.0
    elif j - i == 1: return -1.0
    else: return 0.0
```

```python
# Оформление матрицы, l = 50
h = [[ff(i, j, 50) for j in range(size)] for i in range(size)]
h = np.array(h)
print(h)
```

```python
#  Compute the eigenvalues and right eigenvectors of a square array.
eig, eve = np.linalg.eig(h)
```

```python
eig
```

```python
plt.figure(figsize=(15, 10))
plt.scatter([i for i in range(size)], sorted(eig), s=3)
plt.ylabel('E_n', fontsize=24)
plt.xlabel('n', fontsize=24)
plt.axhline(y = 0, color='r')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot([i for i in range(size)], eve.T[0]**2)
plt.ylabel('Psi_n*Psi_n', fontsize=24)
plt.xlabel('n', fontsize=24)
plt.show()
```

```python
'''(*Пример существования двух локальных уровней:*)'''

size = 1000

def pot2(n):
    if n == 450: 
        return -10
    elif n == 550:
        return -5
    else: 
        return 0
    
    
def f2(i, j):
    if i == j: 
        return 2.0 + pot2(i)
    elif i - j == 1: 
        return -1.0
    elif j - i == 1: 
        return -1.0
    else:
        return 0.0
```

```python
# Оформление матрицы
h = [[f2(i, j) for j in range(size)] for i in range(size)]
h = np.array(h)
print(h)
```

```python
#  Compute the eigenvalues and right eigenvectors of a square array.
eig, eve = np.linalg.eig(h)
```

```python
plt.figure(figsize=(15, 10))
plt.scatter([i for i in range(size)], sorted(eig), s=3)
plt.ylabel('E_n', fontsize=24)
plt.xlabel('n', fontsize=24)
plt.axhline(y = 0, color="r")
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot([i for i in range(size)], eve.T[0]**2, eve.T[1]**2)
plt.ylabel('Psi_n*Psi_n', fontsize=24)
plt.xlabel('n', fontsize=24)
plt.show()
```

```python
'''(*Пример формирования уровней в потенциальной яме:*)'''

size = 1000

def pot3(n):
    if abs(n - 500) <= 20: 
        return -4.0
    else: 
        return 0
    
    
def f3(i, j):
    if i == j: 
        return 2.0 + pot3(i)
    elif i - j == 1: 
        return -1.0
    elif j - i == 1: 
        return -1.0
    else:
        return 0.0
```

```python
# Оформление матрицы
h = [[f3(i, j) for j in range(size)] for i in range(size)]
h = np.array(h)
print(h)
```

```python
#  Compute the eigenvalues and right eigenvectors of a square array.
eig, eve = np.linalg.eig(h)
```

```python
plt.figure(figsize=(15, 10))
plt.plot([i for i in range(size)], [pot3(i) for i in range(size)])
plt.ylabel('V_n', fontsize=24)
plt.xlabel('n', fontsize=24)
#plt.axhline(y = 0, color="r")
plt.show()
```
